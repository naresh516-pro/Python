### Matlplotlib:
-  It is used for Data visualization.
 - mat -> mathematical data
 - plot -> charts/graphs
 - lib -> library
 - first we need to import matplotlib
 - By usin pyplot module we can draw the graphs 
 - Syntax:
           - from matplotlib import pyplot as plt

```python
#import matplotlib library
from matplotlib import pyplot as plt
```

```python
#simple line graph..
x =[10,20,30,40,50]
y=[20,30,40,50,60]
plt.plot(x,y)
plt.show()
```

```python
# apply x ,y label
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#title
plt.title("simple line graph")
plt.show()
```

```python
help(plt.plot)
```

```python
#apply styles,markers,colors...
#line style->ls
#colors->c
#markersmm
#marker face color->mfc
#marker edge color ->mec
#line width -> lw
#marker size -> ms
plt.plot(x,y,c="green")
plt.show()
```

```python
x1 = [10,20,30,40]
y1 = [5,10,15,20]
plt.plot(x1,y1,c="g", marker = "d", mfc = "y", mec = "r", ms = 20,lw = 10)
plt.grid() # To add the grid lines
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("simple graph")
```

```python
#multiple line graphs.
#legend function.
x=[1,2,3,4]
y1=[1,3,4,5]
y2=[4,5,6,7]
plt.plot(x,y1,label='squares')
plt.plot(x,y2,label='cubes')
plt.legend(loc='upper right')
plt.show()
```

### Bar Graph
- By using bar() function we can develop the bar graph 
syntax:
plt.bar(categories,array_values)
- Two types of bar graphs.
- 1.vertical bar graph-> bar()
- 2.Horizontal bar graph -> barh()


```python
# example on bar graph..
subjects=['java','python','dbms','flat']
marks=[20,30,40,50]
plt.bar(subjects,marks,color=['g','r','y','c'],edgecolor="b")
```

```python
# example on horozontal bar graph..
subjects=['java','python','dbms','flat']
marks=[20,30,40,50]
plt.barh(subjects,marks,color=['g','r','y','c'],edgecolor="b")
```

### Scatter Plot
- By using scatter () function we can the scatter plot.
- plt.scatter(x,y)
- it is represented by dot(.)


```python
#Example of Scatter plot.
x=[1,3,4,5,6,7,8]
y=[5,3,6,7,8,6,5]
plt.scatter(x,y,color='g')
plt.grid()
```

### Pie Chart:
- plt.pie(x,labels=names)


```python
#example on pie chart.
names=['a','b','c','d']
per=[23,45,67,64]
plt.pie(per,labels=names)
plt.legend(title='-  Activities')
plt.show()
```
